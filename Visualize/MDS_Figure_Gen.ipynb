```python
import os
import warnings
import matplotlib
import numpy as np
import pandas as pd
import scipy.io as spio
import matplotlib.lines
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
from scipy.stats import spearmanr
from matplotlib.lines import Line2D

def getCueTrialInds(trialType):
    trialInds = []
    startType = []
    goalType = []
    for tr in range(np.shape(trialType)[0]):
        if 'Cue' in trialType[tr,1]:
            trialInds.append(tr)
        if int(trialType[tr,0][0]) == 1:
            goalType.append(trialType[tr,4][0])
        elif int(trialType[tr,0][0]) == 0:
            if goalType.append(trialType[tr,4][0]) == 'E':
                goalType.append('W')
            elif goalType.append(trialType[tr,4][0]) == 'W':
                goalType.append('E')
        startType.append(trialType[tr,3][0])
    return startType,goalType

def getBehaviorData(inputStr,numB):
    otherFile = os.path.join('/home/aditya/Extracted_Data/05_Pos_Decode_FR_XY/Cue/01_Post_Digest/Striatum/',inputStr)
    amat = spio.loadmat(otherFile)
    sArms,gArms = getCueTrialInds(amat.get('behData'))
    startLabels = []
    goalLabels = []
    startGoalLabels = []
    for i in range(len(sArms)):
        if sArms[i] == 'N':
            startLabels.append('^')
            if gArms[i] == 'E':
                goalLabels.append('>')
                startGoalLabels.append('o')
            else:
                goalLabels.append('<')
                startGoalLabels.append('P')
        else:
            startLabels.append('v')
            if gArms[i] == 'E':
                goalLabels.append('>')
                startGoalLabels.append('s')
            else:
                goalLabels.append('<')
                startGoalLabels.append('*')
    return startLabels,goalLabels,startGoalLabels

def loadValuesFromFile(fileOfInterest):
    endStr = fileOfInterest.split('/tmp/')[1]
    rmat = spio.loadmat(fileOfInterest)
    hpcB = rmat.get('hpcEmbed')
    pfcB = rmat.get('pfcEmbed')
    vS,vG,vSG = getBehaviorData(endStr,hpcB)
    return hpcB,pfcB,vS,vG,vSG

def plotFromFile(hpc,v1,v2,v3,outFile):
    legElem1 = [Line2D([0],[0],color='red',label='ID'),
                Line2D([0],[0],color='blue',label='R1'),
                Line2D([0],[0],color='green',label='R2'),
                Line2D([0],[0],color='purple',label='R3'),
                Line2D([0],[0],color='white',marker='>',markersize=12,markerfacecolor='black',label='East'),
                Line2D([0],[0],color='white',marker='<',markersize=12,markerfacecolor='black',label='West'),
               ]
    legElem2 = [Line2D([0],[0],color='red',label='ID'),
                Line2D([0],[0],color='blue',label='R1'),
                Line2D([0],[0],color='green',label='R2'),
                Line2D([0],[0],color='purple',label='R3'),
                Line2D([0],[0],color='white',marker='^',markersize=12,markerfacecolor='black',label='North'),
                Line2D([0],[0],color='white',marker='v',markersize=12,markerfacecolor='black',label='South'),
               ]
    legElem3 = [Line2D([0],[0],color='red',label='ID'),
                Line2D([0],[0],color='blue',label='R1'),
                Line2D([0],[0],color='green',label='R2'),
                Line2D([0],[0],color='purple',label='R3'),
                Line2D([0],[0],color='white',marker='o',markersize=12,markerfacecolor='black',label='NE'),
                Line2D([0],[0],color='white',marker='P',markersize=12,markerfacecolor='black',label='NW'),
                Line2D([0],[0],color='white',marker='s',markersize=12,markerfacecolor='black',label='SE'),
                Line2D([0],[0],color='white',marker='*',markersize=15,markerfacecolor='black',label='SW'),
               ]
    fig = plt.figure(figsize=(24,24))
    ax1 = fig.add_subplot(3,3,1)
    ax2 = fig.add_subplot(3,3,2)
    ax3 = fig.add_subplot(3,3,3)
    for i in range(58,len(v1)):
        ax1.scatter(hpc[i,0],hpc[i,1],marker=v1[i],s=80,color='black')
        ax2.scatter(hpc[i,0],hpc[i,2],marker=v2[i],s=80,color='black')
        ax3.scatter(hpc[i,1],hpc[i,2],marker=v3[i],s=80,color='black')
    ax1.plot(hpc[58:69,0],hpc[58:69,1],color='red')
    ax1.plot(hpc[69:85,0],hpc[69:85,1],color='blue')
    ax1.plot(hpc[85:100,0],hpc[85:100,1],color='green')
    ax1.plot(hpc[100:,0],hpc[100:,1],color='purple')
    ax1.set_xlabel('MDS 1',fontsize=14)
    ax1.set_ylabel('MDS 2',fontsize=14)
    ax1.legend(handles=legElem1,fontsize=12,loc='upper right')
    ax1.set_xticklabels('')
    ax1.set_yticklabels('')

    ax2.plot(hpc[58:69,0],hpc[58:69,2],color='red')
    ax2.plot(hpc[69:85,0],hpc[69:85,2],color='blue')
    ax2.plot(hpc[85:100,0],hpc[85:100,2],color='green')
    ax2.plot(hpc[100:,0],hpc[100:,2],color='purple')
    ax2.set_xlabel('MDS 1',fontsize=14)
    ax2.set_ylabel('MDS 3',fontsize=14)
    ax2.legend(handles=legElem2,fontsize=12,loc='upper right')
    ax2.set_xticklabels('')
    ax2.set_yticklabels('')

    ax3.plot(hpc[58:69,1],hpc[58:69,2],color='red')
    ax3.plot(hpc[69:85,1],hpc[69:85,2],color='blue')
    ax3.plot(hpc[85:100,1],hpc[85:100,2],color='green')
    ax3.plot(hpc[100:,1],hpc[100:,2],color='purple')
    ax3.set_xlabel('MDS 2',fontsize=14)
    ax3.set_ylabel('MDS 3',fontsize=14)
    ax3.legend(handles=legElem3,fontsize=12,loc='upper center')
    ax3.set_xticklabels('')
    ax3.set_yticklabels('')
    #plt.show()
    fig.savefig(outFile,bbox_inches='tight')
    plt.close()
    return

rawF = ''
saF1 = ''
saF2 = ''
blockHPC,blockPFC,labelsG,labelsS,labelsSG = loadValuesFromFile(rawF)
var1 = plotFromFile(blockHPC,labelsS,labelsG,labelsSG,saF1)
var2 = plotFromFile(blockPFC,labelsS,labelsG,labelsSG,saF2)
print('Done')
```
